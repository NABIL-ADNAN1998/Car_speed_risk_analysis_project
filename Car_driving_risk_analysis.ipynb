```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/colab_practice_datasets/car driving risk analysis.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x= df[['speed']]
y=df['risk']
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y, color="red")
plt.xlabel("speed")
plt.ylabel("risk")
plt.title("risk rate by speed")
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state = 1)

```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
reg.predict(x_test)
```

```python
plt.scatter(x,y, color="red")
plt.xlabel("speed")
plt.ylabel("risk")
plt.title("risk rate by speed")
plt.plot(df.speed, reg.predict(df[['speed']]))
```

```python
reg.predict([[80]])
```

```python
reg.score(x_test, y_test)
```
